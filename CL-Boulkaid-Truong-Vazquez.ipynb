```python
def poisson_proc(mu, T):
    num_events = np.random.poisson(mu * T)
    event_times = np.sort(np.random.uniform(0, T, num_events))
    inter_arrival_times = np.diff(event_times)
    return num_events, event_times, inter_arrival_times
```

```python
Nt, Ti, Wi = poisson_proc(mu, 50)
Y = (np.random.exponential(1/lam, Nt))
somme=0
for i in range(len(Nt)):
    Rt_haut = u + p*Ti
    somme = somme + Y[i]
    Rt_bas = u + p*Ti - somme
```
